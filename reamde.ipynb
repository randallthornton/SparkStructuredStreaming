## Troubleshooting

### Header is too large when accessing spark master at http://localhost:9090

This happened because I had additional cookies from my local development efforts. Clearing my cookies for localhost fixed this

```python
# Get Spark Master's local IP via logs

$sparkMaster = "spark-spark-master-1"

docker logs $sparkMaster
```

```python
# Copy program file to spark master container

docker cp -L sum.py "$($sparkMaster):/opt/bitnami/spark/sum.py"
```

```python
# Execute program file via spark-sumit

docker compose exec spark-spark-master-1 spark-submit --master spark://172.21.0.2:7077 sum.py
```
